```python
data = pd.read_csv('Telco_Customer_Churn_Dataset .csv')
data.head()
```

```python
print(data.isnull().sum())
```

```python
data.fillna(data.mean(), inplace=True)
```

```python
print(data[['customerID','gender', 'SeniorCitizen', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod']].columns)
```

```python
categorical_columns = data[['customerID','gender', 'SeniorCitizen', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod']].columns
```

```python
encoder = OneHotEncoder(sparse=False)
```

```python
encoder.fit(data[categorical_columns])
```

```python
encoded_data = encoder.transform(data[categorical_columns])
```
